```python
import matplotlib.pyplot as plt

# Data
languages = [ "SQL", "Power BI", "Python","Excel"]
ratings = [9, 8, 7, 6]

# Create the bar chart
plt.figure(figsize=(8, 4))
plt.bar(languages, ratings, color=["blue", "green", "orange", "red"])
plt.ylim(0, 10)
plt.title("Proficiency Ratings")
plt.ylabel("Rating (Out of 10)")
plt.xlabel("Tools & Languages")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Save the chart
plt.savefig("bar_chart.png")
plt.show()
```
